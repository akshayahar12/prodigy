```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_739890.csv')
df.head()
```

```python
country_population=df.groupby('Country Name')['1960'].sum().reset_index()
print(country_population)
```

```python

country_population=country_population.sort_values('1960',ascending=False)
print(country_population)
```

```python
top_10_countries=country_population.head(10)
print(top_10_countries)
```

```python
plt.bar(top_10_countries['Country Name'],top_10_countries['1960'])
plt.xlabel('country')
plt.ylabel('population(thousands)')
plt.title('top 10 countries by population')
plt.show()
```
